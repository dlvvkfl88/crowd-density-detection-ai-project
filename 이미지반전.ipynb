```python
import cv2
import json
import os

# 이미지 파일이 저장된 디렉토리 경로
image_dir = "/content/images"  # 이미지 디렉토리 경로를 적절하게 수정하세요

# JSON 파일이 저장된 디렉토리 경로
json_dir = "/content/jjson"  # JSON 디렉토리 경로를 적절하게 수정하세요

# 디렉토리 내의 모든 이미지 파일 가져오기
image_files = [f for f in os.listdir(image_dir) if f.endswith((".jpg", ".jpeg"))]

# 이미지 파일 이름의 4자리 부분을 추출
image_names = [os.path.splitext(f)[0][:4] for f in image_files]

# 디렉토리 내의 모든 JSON 파일 가져오기
json_files = [f for f in os.listdir(json_dir) if f.endswith(".json")]

# 이미지 파일 이름과 일치하는 JSON 파일 찾아서 좌우 반전 후 새로운 JSON 생성
for image_name, image_file in zip(image_names, image_files):
    matching_json_file = None

    # 이미지 파일 이름과 일치하는 JSON 파일 찾기
    for json_file in json_files:
        if json_file.startswith(image_name):
            matching_json_file = json_file
            break

    if matching_json_file:
        image_path = os.path.join(image_dir, image_file)
        json_path = os.path.join(json_dir, matching_json_file)

        # 이미지 너비(width) 추출
        image = cv2.imread(image_path)
        if image is not None:
            image_width = image.shape[1]

            # JSON 파일 읽기
            with open(json_path, "r") as json_file:
                data = json.load(json_file)

            # Points 좌우 반전
            flipped_points = []
            for point in data["points"]:
                x, y = point
                flipped_x = image_width - x
                flipped_points.append([flipped_x, y])

            # Boxes 좌우 반전
            flipped_boxes = []
            for box in data["boxes"]:
                x1, y1, x2, y2 = box
                flipped_x1 = image_width - x2
                flipped_x2 = image_width - x1
                flipped_boxes.append([flipped_x1, y1, flipped_x2, y2])

            # 좌우 반전된 데이터를 새로운 JSON으로 저장
            flipped_data = {
                "img_id": data["img_id"],
                "human_num": data["human_num"],
                "points": flipped_points,
                "boxes": flipped_boxes
            }

            # 새로운 JSON 파일 생성
            new_json_path = os.path.join(json_dir, f"{image_name}.json")
            with open(new_json_path, "w") as new_json_file:
                json.dump(flipped_data, new_json_file)
```

```python
import cv2
import json
import os

# 이미지 파일과 JSON 파일의 목록
image_files = ['/content/0001.jpg', '/content/0002.jpg']
json_files = ['/content/0001.json','/content/0002.json']

# 이미지 파일의 너비(width) 추출하는 함수 (여기에서는 임시로 너비를 지정하였습니다)
def extract_image_width(image_files_dir):
    """
    주어진 디렉토리에서 모든 JPG 이미지 파일의 너비를 추출하는 함수

    :param image_files_dir: JPG 이미지 파일들이 저장된 디렉토리 경로
    :return: 이미지 너비를 파일 이름과 함께 딕셔너리로 반환
    """
    image_widths = {}

    # 디렉토리에서 모든 파일 목록 가져오기
    for filename in os.listdir(image_files_dir):
        if filename.endswith(".jpg"):
            # 이미지 파일 경로
            image_path = os.path.join(image_files_dir, filename)

            # 이미지 로드
            image = cv2.imread(image_path)

            # 이미지 너비 추출
            if image is not None:
                height, width, _ = image.shape
                image_widths[filename] = width

    return image_widths

# 이미지 파일의 너비를 추출하여 points와 boxes를 좌우 반전하고 새로운 JSON 파일 생성
for image_file, json_file in zip(image_files, json_files):
    image_width = extract_image_width(image_file)

    image_path = os.path.join('/content', image_file)
    json_path = os.path.join('/content', json_file)

    # 이미지 너비(width) 추출
    image = cv2.imread(image_path)
    if image is not None:
        # JSON 파일 읽기
        with open(json_path, "r") as json_file:
            data = json.load(json_file)

        # Points 좌우 반전
        flipped_points = []
        for point in data["points"]:
            x, y = point
            flipped_x = image_width - x
            flipped_points.append([flipped_x, y])

        # Boxes 좌우 반전
        flipped_boxes = []
        for box in data["boxes"]:
            x1, y1, x2, y2 = box
            flipped_x1 = image_width - x2
            flipped_x2 = image_width - x1
            flipped_boxes.append([flipped_x1, y1, flipped_x2, y2])

        # 좌우 반전된 데이터를 새로운 JSON으로 저장
        flipped_data = {
            "img_id": data["img_id"],
            "human_num": data["human_num"],
            "points": flipped_points,
            "boxes": flipped_boxes
        }

        # 새로운 JSON 파일 생성
        new_json_path = os.path.join('/content', f"flipped_{json_file}")
        with open(new_json_path, "w") as new_json_file:
            json.dump(flipped_data, new_json_file)

print("좌우 반전된 JSON 파일 생성 완료")
```

```python
import json
import os
import cv2

# 디렉토리 경로 설정
image_dir = "/content/"  # 이미지 디렉토리 경로를 적절하게 수정하세요
json_dir = "/content/"  # JSON 디렉토리 경로를 적절하게 수정하세요
output_json_dir = "/content/flipped_json"  # 새로운 JSON 파일이 저장될 디렉토리 경로를 수정하세요

# 디렉토리 내의 모든 이미지 파일 가져오기
image_files = [f for f in os.listdir(image_dir) if f.endswith((".jpg", ".png", ".jpeg"))]

# 이미지 파일 별로 폭(width) 추출
image_widths = {}

for image_file in image_files:
    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)
    if image is not None:
        image_widths[image_file] = image.shape[1]

# JSON 파일 경로 목록
json_file_paths = [os.path.join(json_dir, f"{os.path.splitext(image_file)[0]}.json") for image_file in image_files]

for json_file_path, image_file in zip(json_file_paths, image_files):
    # JSON 파일을 읽어옵니다.
    with open(json_file_path, "r") as json_file:
        data = json.load(json_file)

    # 해당 이미지 파일의 너비 추출
    image_width = image_widths[image_file]

    # Points 좌우 반전
    flipped_points = []
    for point in data["points"]:
        x, y = point
        flipped_x = image_width - x
        flipped_points.append([flipped_x, y])

    # Boxes 좌우 반전
    flipped_boxes = []
    for box in data["boxes"]:
        x1, y1, x2, y2 = box
        flipped_x1 = image_width - x2
        flipped_x2 = image_width - x1
        flipped_boxes.append([flipped_x1, y1, flipped_x2, y2])

    # 좌우 반전된 데이터를 새로운 JSON으로 저장
    flipped_data = {
        "img_id": data["img_id"],
        "human_num": data["human_num"],
        "points": flipped_points,
        "boxes": flipped_boxes
    }

    # 새로운 JSON 파일 경로 설정
    os.makedirs(output_json_dir, exist_ok=True)
    output_json_path = os.path.join(output_json_dir, f"r_{os.path.basename(json_file_path)}")

    # JSON 데이터를 파일로 저장
    with open(output_json_path, "w") as json_file:
        json.dump(flipped_data, json_file, indent=4)

    # 결과 출력
    print(f"좌우 반전된 Points (JSON 파일: {json_file_path}):")
    print(flipped_points)
    print(f"\n좌우 반전된 Boxes (JSON 파일: {json_file_path}):")
    print(flipped_boxes)
    print("\n새로운 JSON 파일 경로:")
    print(output_json_path)
    print("\n" + "=" * 50 + "\n")
```

```python
import json
import cv2
import os

# 디렉토리 경로 설정
json_dir = "/content/jjson"  # JSON 파일이 있는 디렉토리 경로
image_dir = "/content/images"  # 이미지 파일이 있는 디렉토리 경로
output_dir = "/content/output"  # 결과를 저장할 디렉토리 경로

# JSON 파일과 이미지 파일 목록 가져오기
json_files = [f for f in os.listdir(json_dir) if f.endswith(".json")]
image_files = [f for f in os.listdir(image_dir) if f.endswith((".jpg", ".png", ".jpeg"))]

# 디렉토리 생성 (결과를 저장할 디렉토리)
os.makedirs(output_dir, exist_ok=True)

# JSON 파일을 처리하는 반복문
for json_file in json_files:
    json_file_path = os.path.join(json_dir, json_file)

    # JSON 파일을 읽어옵니다.
    with open(json_file_path, "r") as json_file:
        data = json.load(json_file)

    # 이미지 파일 경로
    base_name = os.path.splitext(json_file.name)[0]  # JSON 파일의 기본 이름
    image_file_name = base_name + ".jpg"  # JSON 파일과 이미지 파일 이름이 일치해야 합니다.
    image_file_path = os.path.join(image_dir, image_file_name)

    # 이미지를 읽어옵니다.
    image = cv2.imread(image_file_path)

    # 이미지 폭 (너비) 추출
    if image is not None:
        image_width = image.shape[1]
    else:
        # 이미지를 읽을 수 없는 경우에 대한 처리
        continue

    # Points 좌우 반전
    flipped_points = []
    for point in data["points"]:
        x, y = point
        flipped_x = image_width - x
        flipped_points.append([flipped_x, y])

    # Boxes 좌우 반전
    flipped_boxes = []
    for box in data["boxes"]:
        x1, y1, x2, y2 = box
        flipped_x1 = image_width - x2
        flipped_x2 = image_width - x1
        flipped_boxes.append([flipped_x1, y1, flipped_x2, y2])

    # 좌우 반전된 데이터를 새로운 JSON으로 저장
    flipped_data = {
        "img_id": data["img_id"],
        "human_num": data["human_num"],
        "points": flipped_points,
        "boxes": flipped_boxes
    }

    # 결과 JSON 파일 경로 설정
    output_json_path = os.path.join(output_dir, "r_" + json_file)

    # JSON 데이터를 파일로 저장
    with open(output_json_path, "w") as output_json_file:
        json.dump(flipped_data, output_json_file, indent=4)

    # 결과 출력
    print(f"좌우 반전된 JSON 파일 생성: {output_json_path}")
```
